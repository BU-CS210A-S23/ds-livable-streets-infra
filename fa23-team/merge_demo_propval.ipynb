# Join demographics.csv and property_value.csv


```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')
data_folder = "/content/drive/MyDrive/DS701/"
```

```python
# Read dfs
demo = pd.read_csv(data_folder + "demographics.csv", index_col=False)
prop_val = pd.read_csv(data_folder + "property_value.csv", index_col=False)

print(demo.shape)
print(prop_val.shape)
```

```python
# Merge prop_val and demo
merged_df = pd.merge(demo, prop_val, on=['tract', 'year'])

merged_df.head()
```

```python
# Save merged_df
merged_df.to_csv(data_folder + "combined_demo_propval.csv", index=False)
```
