```python
# Read 2021 ACS Demographic Data
import pandas as pd

data = pd.read_csv(data_folder+'2021_demographic.csv', index_col=False)
data.head()
```

```python
data = data.stack(0)
data = data.unstack()
```

```python
data
```

```python
# # Remove NaN headers
# cap_filter = data['Label (Grouping)'].str.isupper()
# data = data[~cap_filter]

# data.head()
```

```python
# Transpose data
# cols = data['Label (Grouping)']

data = data.transpose()

data.head()
```

```python
# Extract the first row as column labels
new_columns = data.iloc[0]

# Set the first row as the column labels
data = data.iloc[1:]
data = data.set_axis(new_columns, axis=1)

# Reset the index
data = data.reset_index()
data.columns.names = ['']
```

```python
data = data.rename(columns= {"index":"Tract"})
data.head()
```

```python
data[['Tract', 'Info']] = data['Tract'].str.split(',', 1, expand=True)
data.head()
```

```python
data['Info'] = data['Info'].str.replace('Suffolk County, Massachusetts!!', '', regex=True)
data['Tract'] = data['Tract'].str.replace('Census Tract ', '', regex=True)
data.head()
```

```python
data.columns
```

```python
data.columns = data.columns.str.strip()
data.columns
```

```python
cols_to_keep = ['Tract', 'Info', 'Male', 'Female', 'Median age (years)',  ' White',
       'Black or African American',
       'American Indian and Alaska Native', 'Asian',
       'Native Hawaiian and Other Pacific Islander',
       'Some other race', 'Hispanic or Latino (of any race)']

data_clean = data.drop([col for col in data.columns if col not in cols_to_keep], axis=1)
```

```python
data_clean
```

```python
data_clean.Info = data_clean.Info.str.strip()
```

```python
# List of strings to check for in the rows
strings_to_remove = ["Margin of Error", "Percent", "Percent Margin of Error"]

data_clean = data_clean[~data_clean['Info'].isin(strings_to_remove)]
```

```python
data_clean
```

```python
# Create an empty set to track seen column names
seen_columns = set()

unique_columns = []

# Iterate through columns and keep the first occurrence
for column in data_clean.columns:
    if column in cols_to_keep and column not in seen_columns:
        unique_columns.append(column)
        seen_columns.add(column)

# Select and create a DataFrame with unique columns
#data_final = data_clean[unique_columns]
```
