```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('eviction.csv')
violation_type_column = df['violation_type']
print(violation_type_column)
```

```python
violation_counts = df['violation_type'].value_counts()

# This is for plotting.
violation_counts.plot(kind='bar')
plt.title('Violation Type Frequencies')
plt.xlabel('Violation Type')
plt.ylabel('Frequency')
plt.xticks(rotation=90) 
plt.show()
```

```python
# Here, we are grouping by neighborhood and violation type and counting the occurrences
grouped_data = df.groupby(['neighborhood', 'violation_type']).size().unstack(fill_value=0)

grouped_data.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Violation Types by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Count of Violation Types')
plt.xticks(rotation=90)  
plt.legend(title='Violation Type')
plt.show()
```
